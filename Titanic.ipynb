# Data Cleaning - Titanic Dataset

```python
import kagglehub
import os
import pandas as pd

path = kagglehub.dataset_download("yasserh/titanic-dataset")
print("Path of data set", path)

print(os.listdir(path))
file = os.path.join(path, 'Titanic-Dataset.csv')

data = pd.read_csv(file)

data.head()
```

```python
data.describe()
```

```python
print(data.isnull())
```

```python
print(data.isnull().sum())
```

```python
data['Age'] = data['Age'].fillna('Unknown')
```

```python
data['Cabin'] = data['Cabin'].fillna('Unknown')
```

```python
data.isnull().sum()
```

## Label Encoding

```python
data['Sex'] = data['Sex'].astype(str).str.strip().str.lower()
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
print(data['Sex'].head())
```

```python
print(data['Embarked'].unique())
```

## one hot encoding

```python
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data['Embarked'].unique()
```

```python
data = pd.get_dummies(data, columns = ['Embarked'], prefix=['Embarked'])
data.head()
```

```python
data[['Embarked_C','Embarked_Q','Embarked_S']] = data[['Embarked_C','Embarked_Q','Embarked_S']].astype(int)
data.head()
```

```python
data['Cabin'].unique()
```

```python
data['Deck'] = data['Cabin'].apply(lambda x: x[0] if x != 'Unknown' else 'U') 
print(data['Deck'].head())
```
